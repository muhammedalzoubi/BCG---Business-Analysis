```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Sector": ["Sector A", "Sector B", "Sector C"],
    "Capital_Expenditure": [8_000_000, 8_000_000, 8_000_000,],
    "Net_Income_Year_4": [715_000, 292_500, 65_000],
    "Annual_Cash_Flow": [715_000, 292_500, 65_000],
    "EBITDA_Year_1": [1_100_000, 1_100_000, 450_000]
}

df = pd.DataFrame(data)

df["ROI_Year_4_%"] = (df["Net_Income_Year_4"] / df["Capital_Expenditure"]) * 100

df["Payback_Period_Years"] = (
    df["Capital_Expenditure"] / df["Annual_Cash_Flow"]
)

results = df[[
    "Sector",
    "ROI_Year_4_%",
    "Payback_Period_Years",
    "EBITDA_Year_1"
]].round(2)

results
```

```python
import matplotlib.pyplot as plt

def add_labels(bars, format_type="number"):
    for bar in bars:
        value = bar.get_height()
        if format_type == "percent":
            label = f"{value:.1f}%"
        elif format_type == "years":
            label = f"{value:.2f} yrs"
        else:
            label = f"{value:,.0f}"

        plt.text(
            bar.get_x() + bar.get_width() / 2,
            value,
            label,
            ha='center',
            va='bottom',
            fontsize=10
        )

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Financial Performance by Sector", fontsize=16, fontweight="bold")

bars1 = axes[0].bar(df["Sector"], df["ROI_Year_4_%"])
axes[0].set_title("ROI – Year 4")
axes[0].set_ylabel("ROI (%)")
axes[0].grid(axis='y', linestyle='--', alpha=0.6)
plt.sca(axes[0])
add_labels(bars1, "percent")

bars2 = axes[1].bar(df["Sector"], df["Payback_Period_Years"])
axes[1].set_title("Payback Period")
axes[1].set_ylabel("Years")
axes[1].grid(axis='y', linestyle='--', alpha=0.6)
plt.sca(axes[1])
add_labels(bars2, "years")

bars3 = axes[2].bar(df["Sector"], df["EBITDA_Year_1"])
axes[2].set_title("EBITDA – Year 1")
axes[2].set_ylabel("Value")
axes[2].grid(axis='y', linestyle='--', alpha=0.6)
plt.sca(axes[2])
add_labels(bars3)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```
